# Iris dataset classifier

```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
```

## 1. Data wrangling & cleaning

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
iris.keys()
```

```python
print(iris['DESCR'])
```

```python
print("FEATURE NAMES: {}".format(iris['feature_names']))
```

```python
print("TARGET NAMES: {}".format(iris['target_names']))
```

## 2. Feature engineering

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
data_points = iris['data']
labels = iris['target']
```

```python
print(data_points[0])
```

### Split dataset into Train and Test

```python
x_train, x_test, y_train, y_test = train_test_split(data_points, labels, test_size=0.3, random_state=50)
```

```python
print("Data set size: Train set: {0} flowers, Test set: {1} flowers".format(len(x_train), len(x_test)))
```

### Normalize data

```python
scaler = MinMaxScaler()
```

```python
x_train_normalized = scaler.fit_transform(x_train)
x_test_normalized = scaler.transform(x_test)
```

## 3. Training

### Create model

```python
from tensorflow.keras import models
from tensorflow.keras import layers
```

```python
model = keras.models.Sequential()
```

### Add Input, Hidden and Output layers

```python
# Input layer
model.add(layers.Dense(units=10, input_dim=4, activation='relu'))

# Hidden layer(s)
model.add(layers.Dense(units=10, activation='relu'))

# Output layer
model.add(layers.Dense(units=3, activation='softmax'))
```

### Add training parameters

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

### Start training on train set

```python
model.fit(x_train_normalized, y_train, epochs=1000)
```

## 4. Evaluation

```python
from sklearn.metrics import classification_report
```

### Predict entire test set

```python
predictions = model.predict_classes(x_test_normalized)
```

### Visualize result

```python
print(classification_report(predictions, y_test))
```
